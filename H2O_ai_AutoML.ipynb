# H2O.ai AutoML

```python
import h2o
from h2o.automl import H2OAutoML
```

```python
h2o.init()
```

Read csv

```python
import pandas as pd
import numpy as np
y_tr = pd.read_csv('y_tr_06.csv', dtype={'time_to_failure': np.float64}, index_col=0)
X_tr = pd.read_csv('X_tr_06.csv', index_col=0)
```

```python
X_tr['time_to_failure']=y_tr['time_to_failure']
X_tr.head()
```

```python
X_tr.to_csv('train_06.csv')
```

```python
train = h2o.import_file("train_06.csv")
test = h2o.import_file("X_test_06.csv")
```

Define target and features

```python
train.head()
```

```python
exclude_cols = ["C1", "time_to_failure"]
x = [colnames for colnames in train.col_names if colnames not in exclude_cols]
```

```python
y = "time_to_failure"
```

```python
test.head()
```

AutoML

```python
aml = H2OAutoML(max_models=20, seed=1)
aml.train(x=x, y=y, training_frame=train)
```

Leader Board

```python
# View the AutoML Leaderboard
lb = aml.leaderboard
lb.head(rows=lb.nrows)  # Print all rows instead of default (10 rows)
```

Confirm best model

```python
aml.leader
```

Predict

```python
preds = aml.predict(test)
```

```python
preds
```

```python
test["time_to_failure"] = preds
```

```python
h2o.export_file(test[["C1", "time_to_failure"]], path="submission_06.csv")
```
